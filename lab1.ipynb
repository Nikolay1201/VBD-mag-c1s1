```python
price_bins = [0, 100000, 200000, 300000, 400000, 500000, 600000]
price_labels = ['0-100k', '100k-200k', '200k-300k', '300k-400k', '400k-500k', '500k-600k']
# Генерация временного признака для сводной таблицы
df['PriceRange'] = pd.cut(df['Price'], bins=price_bins, labels=price_labels)
pd.pivot_table(data=df, values='Price', index='Rooms', columns='PriceRange', aggfunc='mean')
df = df.drop(['PriceRange'], axis=1)
```

```python
# В данном случае поскольку фичи количественные считаю замену на медиану достаточным
df['LifeSquare'].fillna(df['LifeSquare'].median(), inplace=True);
df['Healthcare_1'].fillna(df['Healthcare_1'].median(), inplace=True);
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(df['Price'], bins=30, color='blue', edgecolor='black')
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(df['Square'], df['Price'], alpha=0.5)
plt.title('Square Footage vs. Price')
plt.xlabel('Square Footage')
plt.ylabel('Price')
plt.show()
```
